```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

Vamos a generar intervalo de fechas

```python
anos = [str(i) for i in range(2004, 2008)]
meses = ['01', '02', '03', '04', '05', '06',
              '07', '08', '09', '10', '11', '12']
fechas = []
for i in range(len(anos)):
    for j in range(len(meses)):
        for k in range(1, 30):
            if k < 10:
                fecha_i = f"{anos[i]}-{meses[j]}-0{k}-12"
                #print(fecha_i)
            else:
                fecha_i = f"{anos[i]}-{meses[j]}-{k}-12"
                #print(fecha_i)
            fechas.append(fecha_i)
            
```

Definamos promedio (de Sebastian)

```python
def promedio(imag):
    """Funcion para promediar. suma valor de cada pixel y divide en numero de pixeles"""
    x, y = imag.shape
    suma = 0
    div = x*y
    for i in range(0, x):
        for j in range(0, y):
            suma = suma + imag[i, j]
    prom = suma/div
    
    return prom
```

Podemos rodar para varias fechas (aqui 500 dias) y hacer un grÃ¡fico

```python
tierra_12 = [0*i for i in range(len(fechas))]
escala_de_gris_tierra_12 = [0*i for i in range(len(fechas))]
vapor_de_agua_promedio = [0*i for i in range(len(fechas))]

dias_no_disponibles = 0

for dia in range(1,500):
    try:
        tierra_12[dia] = np.array(cv2.imread(fechas[dia]))
        color_a_grises = cv2.cvtColor(tierra_12[dia], cv2.COLOR_BGR2GRAY)
        grises_normalizado = cv2.normalize(color_a_grises, None, alpha = 0, beta = 100, norm_type = cv2.NORM_MINMAX, dtype = cv2.CV_32F)
        vapor_de_agua_promedio[dia] = promedio(grises_normalizado)


    except:
        dias_no_disponibles += 1
    
print (f"{dias_no_disponibles/len(fechas)}")
```

```python
plt.plot( fechas[1:500], vapor_de_agua_promedio[1:500])
```
