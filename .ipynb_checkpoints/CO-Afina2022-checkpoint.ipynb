## Hackaton Co-Afina2022: LA-CoNGA physics

* Hicimos uso de bases de datos abiertos [GIBS](https://www.ncdc.noaa.gov/gibbs/year) los cuales fueron proporcionados por los organizadores.
* Proponemos un índice para evaluar la variación del vapor de agua sobre Latinoamérica. Este índice consiste en el promedio de intensidad de la imagen después de ser normalizado a escala de grises. [ver código](https://github.com/Migusb/CO-Afina2022/blob/master/gibs/imagekpi.py)

### Linea de tiempo de la intensidad promedio

```python

fig, ax = timeseriesplots({'Vapor de agua':(df[df['average']>60]/df[df['average']>60].values.min())*100}, ylabel= 'Intensidad promedio (%)')
```
